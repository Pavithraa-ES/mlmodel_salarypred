```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train ,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
#data normalization with sklearn
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
#fit scalar on training set
x_train = sc.fit_transform(x_train)
#transform training db
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#importing machine learning techniques
```

```python
from sklearn.tree import DecisionTreeRegressor
#calling dt with max depth 3
dt = DecisionTreeRegressor(max_depth=3)

```

```python
dt.fit(x_train , y_train)
```

```python
print(dt.score(x_train,y_train))
print(dt.score(x_test,y_test))
```

```python
y_pred=dt.predict(x_test)
print(y_pred)
```

```python
from sklearn import metrics
```

```python
r_square=metrics.r2_score(y_test,y_pred)
print('R-square error associated with decision tree regressor is: ',round(100*(r_square),3))
```

```python
y_pred=dt.predict(x_test)
r_square=metrics.r2_score(y_test,y_pred)
print('R-square error associated with decision tree regressor is: ',round(100*(r_square),2))
```

```python
from sklearn.ensemble import RandomForestRegressor
#calling dt with max depth 3
from sklearn import metrics
rfr = RandomForestRegressor(random_state=5)
rfr.fit(x_train,y_train)
```

```python
y1_pred=rfr.predict(x_test)
```

```python
print("MEAN ABSOLUTE ERROR(MAE):", metrics.mean_absolute_error(y_test,y1_pred))
print("MEAN SQUARED ERROR(MSE):", metrics.mean_squared_error(y_test,y1_pred))
print("ROOT MEAN SQUARED ERROR(RMSE):", np.sqrt(metrics.mean_squared_error(y_test,y1_pred)))

mape= np.mean(np.abs((y_test-y1_pred)/ np.abs(y1_pred)))
mape1=round(mape*100,2)
accuracy1=round(100*(1-mape),2)

print("Mean Absolute Percenatge Error(MAPE):", round(mape*100,2))
print("Accuracy:", round(100*(1-mape),2))
```

```python
#CROSS VALIDATION- it is a resampling method that uses different portions of the data to test and train a model on different iterations
```

```python
from sklearn.model_selection import ShuffleSplit
model= LinearRegression()
ssplit= ShuffleSplit(n_splits=10,test_size=0.30)
from sklearn.model_selection import cross_val_score
results=cross_val_score(model,x,y,cv=ssplit)
print(results)
print(np.mean(results))
```

```python
from sklearn.model_selection import ShuffleSplit
model= DecisionTreeRegressor(max_depth=3)
ssplit= ShuffleSplit(n_splits=10,test_size=0.30)
from sklearn.model_selection import cross_val_score
results=cross_val_score(model,x,y,cv=ssplit)
print(results)
print(np.mean(results))
```
